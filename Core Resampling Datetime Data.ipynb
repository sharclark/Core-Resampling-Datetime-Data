# Part 1 - Load and make a datetime index

```python
import datetime as dt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mticks
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('Data/london_weather_MODIFIED - london_weather_MODIFIED.csv')
df.info()
df.head()
```

```python
df['datetime']= pd.to_datetime(df['date'])
df.info()
df.head()
```

```python
## displaying random sample
df['date'].sample(n=10).sort_index()
```

```python
pd.to_datetime(df['date'], format='%Y%m%d')
```

```python
df = df.drop(columns='datetime')
df.head()
```

```python
df['datetime']= pd.to_datetime(df['date'], format='%Y%m%d')
df.head()
```

```python
#set datetime as index
df = df.set_index('datetime')
df.head()
```

```python
df.index
#no frequency yet
```

```python
#filter out year from 2000 and later
df= df.loc['2000':]
df
```

```python
#keep only precip, mean_temp, min_temp, max_temp, snow_depth
df = df.drop(['date', 'cloud_cover', 'sunshine', 'global_radiation', 'pressure'], axis = 1)
df
```

```python
#impute any missing values
df.isna().sum()
```

```python
df['snow_depth']
```

```python
#snow depth and precip should be forward filled, not interpolated
#temps can be interpolated

###snow_depth
df['snow_depth_ff']= df['snow_depth'].fillna(method='ffill')

df['snow_depth_ff']
```

```python
###precip
df['precipitation']= df['precipitation'].fillna(method='ffill')
df.head()
```

```python
df.isna().sum()
```

```python
### max_temp, mean_temp, and min_temp
df['max_temp']= df['max_temp'].interpolate()
df['mean_temp']= df['mean_temp'].interpolate()
df['min_temp']= df['min_temp'].interpolate()
```

```python
df.isna().sum()
```

```python
df = df.drop(['snow_depth'], axis = 1)
df
```

```python
df = df.rename(columns = {'snow_depth_ff': 'snow_depth'})
df.head()
```

# Q1: What month had the most precipitation between 2000 through 2010?

```python
# make a new df
df2 = df.copy()
df2

```

```python
# filter out just 2000-2010
df2 = df2.loc['2000':'2010']
df2
```

```python
#resample as month using sum for aggregate because looking for total precip for month
df2_month_res = df2.resample('MS').sum()
df2_month_res.name = "Month_Resampled"
df2_month_res
```

```python
max_date = df2_month_res['precipitation'].idxmax()
max_date
```

```python
df2_month_res.loc[max_date]
```

```python
#save variables for print statement
max_temp, mean_temp, min_temp, precipitation, snow_depth = df2_month_res.loc[max_date]
```

```python
## now print the requested statement
print(f"The highest precipitation month was {max_date.strftime('%B')}. The precipitation was {precipitation:.2f}")
```

```python
ax= df2_month_res['precipitation'].plot()
```

```python
plt.rcParams['figure.figsize']
```

```python
df2_month_res['precipitation'].plot(figsize=(12,3));
```

```python
## creating our tick locators and formatters


loc_1_year = mdates.YearLocator()


fmt_year = mdates.DateFormatter("%Y")
```

```python
fig, ax = plt.subplots(figsize=(12,3))
ax = df2_month_res['precipitation'].plot(figsize=(12,3));
ax.set(ylabel='Precipitation', title= "Precipitation for 2000-2010", xlabel = 'Years')
ax.xaxis.set_major_locator(loc_1_year)
ax.xaxis.set_major_formatter(fmt_year)
df2_month_res['precipitation'].plot(ax=ax)
fig.autofmt_xdate(rotation=90,ha='center');
```

# Q2: Which year between 2000-2020 had the coolest average temperature? 

```python
# make a new df
df3 = df.copy()
df3
```

```python
# dont need to filter years, that is already the range
## resample as year and use average as aggregate
df3_year_res = df3.resample('Y').mean()
df3_year_res.name = "Year_Resampled"
df3_year_res
```

```python
min_date = df3_year_res['mean_temp'].idxmin()
min_date
```

```python
df3_year_res.loc[min_date]
```

```python
#save variables for print statement
max_temp, mean_temp, min_temp, precipitation, snow_depth = df3_year_res.loc[min_date]
```

```python
## now print the requested statement
print(f"The lowest average temperature year was {min_date.strftime('%Y')}. The average temperature was {mean_temp:.2f}")
```

```python
fig, ax = plt.subplots(figsize=(12,3))
ax = df3_year_res['mean_temp'].plot(figsize=(12,3));
ax.set(ylabel='Degrees', title= "Average Temperature for 2000-2020", xlabel = 'Years')
ax.xaxis.set_major_locator(loc_1_year)
ax.xaxis.set_major_formatter(fmt_year)
df3_year_res['mean_temp'].plot(ax=ax)
fig.autofmt_xdate(rotation=90,ha='center');
```
